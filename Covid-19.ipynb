```python
Covid_data.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace=True, axis=1)
```

```python
Covid_data.head()
```

```python
Covid_data['Date'] = pd.to_datetime(Covid_data['Date'], format = '%Y-%m-%d')
Covid_data.head()
```

```python
#Active coses
Covid_data['Active_cases'] = Covid_data['Confirmed'] - (Covid_data['Cured'] + Covid_data['Deaths'])
Covid_data.head()
```

```python
statewise = pd.pivot_table(Covid_data, values = ['Confirmed', 'Deaths', 'Cured'], index = 'State/UnionTerritory', aggfunc = max)
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
statewise['Death Rate'] = statewise['Deaths']*100/statewise['Confirmed']
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
# top 10 actibe cases state
top_10_active_cases = Covid_data.groupby(by = 'State/UnionTerritory').max()[['Active_cases']].sort_values(by = 'Active_cases', ascending = False).reset_index()
fig = plt.figure(figsize = (16, 9))
plt.title('Top 10 states with most active cases in India', size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], x = 'State/UnionTerritory', y = 'Active_cases', linewidth = 2, edgecolor = 'red')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, ha = 'right', size = 15)
```

```python
#Top states with hightest death
top_10_deaths = Covid_data.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = 'Deaths', ascending = False).reset_index()
```

```python
fig = plt.figure(figsize =(18, 5))
plt.title('Top 10 states with most deaths in India', size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], x = 'State/UnionTerritory', y = 'Deaths', linewidth = 2, edgecolor = 'black')
plt.xlabel('States', size = 15)
plt.ylabel('Deaths', size = 15)
plt.show()
```

```python
#Growth trend
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot(data = Covid_data[Covid_data['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerela', 'Tamil Nadu', 'Rajasthan', 'Uttar Pradesh'])], x = 'Date', y = 'Active_cases', hue = 'State/UnionTerritory')
ax.set_title('Top 5 Affected states in India', size = 16)
```

```python
#Remove rows where state = India
vaccine = vaccine_data[vaccine_data['State'] != 'India']
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated': 'Total'})
```

```python
#Most vaccinated states
max_vac = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total').sort_values('Total', ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title("Top 5 Vaccination State in India", size= 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
